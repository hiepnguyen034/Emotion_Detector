```python
cascadePath = "C:\\Users\\Hiep Nguyen\\Desktop\\Emotion_Detector\\data\\haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(cascadePath);
```

```python
recognizer = cv2.face.LBPHFaceRecognizer_create()
```

```python
def reshape_input_image(path):
    img = Image.open(path).convert('L')
    img = resizeimage.resize_contain(img, [243, 320])
    image = np.array(img, 'uint8')
    images=[]
    faces = faceCascade.detectMultiScale(image)    
    for (height, weight, w, h) in faces:
        images.append(image[weight: weight + 151, height: height + 151])
    return np.dot(images[0][...,:3], [0.299, 0.587, 0.114])
```
